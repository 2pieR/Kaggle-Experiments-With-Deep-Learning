```python
user_in = Input(shape=(1,),dtype='int64',name='user_in')
u = Embedding(n_users,n_factors,input_length=1,W_regularizer=l2(1e-4))(user_in)
```

```python
movie_in = Input(shape=(1,),dtype='int64',name='movie_in')
m = Embedding(n_movie,n_factors,input_length=1,W_regularizer=l2(1e-4))(movie_in)
```

## Adding Bias

```python
def create_bias(inp,n_in):
    x= Embedding(n_in,1,input_length=1)(inp)
    return Flatten()(x)
```

```python
def create_bias(inp, n_in):
    x = Embedding(n_in, 1, input_length=1)(inp)
    return Flatten()(x)
```

```python
user_bias = create_bias(user_in,n_users)
```

```python
movie_bias
```

```python
movie_bias = create_bias(movie_in,n_movie)
```

```python
x = merge([u,m],mode='dot')
x= Flatten()(x)
x = merge([x,user_bias],mode='sum')
x = merge([x,movie_bias],mode='sum')
model = Model([user_in,movie_in],x)
model.compile(Adam(0.001),loss='mse')
```

```python
model.fit([trn.userId,trn.movieId],trn.rating,batch_size=64,nb_epoch=10,validation_data=([val.userId,val.movieId],val.rating))
```

```python
model.optimizer.lr = 0.01
model.fit([trn.userId,trn.movieId],trn.rating,batch_size=64,nb_epoch=3,validation_data=([val.userId,val.movieId],val.rating))
```

```python
model.optimizer.lr = 0.0001
model.fit([trn.userId,trn.movieId],trn.rating,batch_size=64,nb_epoch=10,validation_data=([val.userId,val.movieId],val.rating))
```
